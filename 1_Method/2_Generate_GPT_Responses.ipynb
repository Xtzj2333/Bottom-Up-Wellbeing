```python
import os
from openai import OpenAI
import time

here = '1_Method/'
```

```python
start_index = 0
end_index = 100
```

```python
in_file = here + 'data/ideal_life_adjectives.pkl'
out_file = here + f'data/DGP/adj_{start_index}_{end_index}.csv'

adjectives = ['beautiful', 'happy']

adjectives = adjectives[start_index:end_index]
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

prompt = """
Please take a moment to reflect on your life as a whole.
Then, think about the ideal life you would like to live.
What is the type of life you want to live?
Please provide a brief response (a couple of sentences).
"""

def assess_comment(adjective):
    system_message = f"""
    There are many adjectives that can describe the life of a person.
    Assume you're a person whose ideal life is a {adjective} life.
    The adjective, {adjective}, describes your ideal life; other adjectives do not.
    A life characterized by other adjectives is not your ideal life.
    Your responses do not have to be formal.
    """
    try:
        response = client.responses.create(
            model="gpt-4o-mini",
            input=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt},
            ],
        )
        return response.output_text
    except Exception as e:
        print(f"Error processing request for adjective '{adjective}': {e}")
        return None
```

```python
from tqdm import tqdm
import pandas as pd
import time

rows = []

for adj in tqdm(adjectives):
    for i in range(2):
        resp = assess_comment(adj)
        rows.append({"adj": adj, "response": resp})
        time.sleep(1)

output_df = pd.DataFrame(rows)
output_df.to_csv(out_file, index=False)
```
